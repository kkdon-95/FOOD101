```python
import pandas as pd
import numpy as np
```

```python
file=pd.read_csv('IITK_Courses_Report.csv')
```

```python
file.isnull()
```

```python
file.isnull().sum()
```

```python
file.dropna(subset=['Student Name','Course'],inplace=True)
```

```python
file
```

```python
file.reset_index(inplace=True)
```

```python
file
```

```python
def geometric_mean(file, column):

    for i in range(1,184):
        if pd.isnull(file.loc[i, column]):
            a = file.loc[i - 1, column]
            b = file.loc[i + 1, column]
            file.loc[i, column] = np.sqrt(a*b)

for column in file.select_dtypes(include=['float64', 'int64']).columns:
    geometric_mean(file, column)
```

```python
file.drop('index',axis=1,inplace=True)
```

```python
file.to_csv('cleaned_IITK_Courses_Report.csv',index=False)
```
